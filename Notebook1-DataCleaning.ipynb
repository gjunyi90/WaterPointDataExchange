## Data Cleaning

```python
import pandas as pd
import numpy as np
from reduce_size import reduce_mem_usage
import matplotlib.pyplot as plt
import re
from collections import Counter

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 1000)
%matplotlib inline
df = pd.read_csv('Water_Point_Data_Exchange_Complete_Dataset (2).csv')
df, NA_List = reduce_mem_usage(df)
```

```python
print(df.describe())
print(df.dtypes)
df.head()
```

```python
df_focus = df[['water_source', 'water_tech', 'status_id','management','installer','country_id']]
```

### Clean 1: Remove Special Characters

```python
# Lower cased
df_focus.loc[:,'water_tech'] = df_focus.loc[:,'water_tech'].str.lower()

# Replace special characters, extra whitespace, empty string
df_focus.loc[:,'water_tech2']=df_focus.loc[:,'water_tech'].str.replace(r'\W+', ' ').replace(r'\s+',' ').replace(r'^\s*$', np.nan, regex=True)

# Find the difference (before and after)
df_focus.loc[:,['water_tech','water_tech2']][(~pd.isnull(df_focus.water_tech)) & (df_focus.water_tech != df_focus.water_tech2)].drop_duplicates()
```

### Identify rows with words that contain special words
How much of these data make up the entire tables?

```python
detect_list = ['other'
            ,'none'
            ,'no'
            ,'0']

# To detect the words
def special_detect(word):
    detected = None
    for alert in detect_list:
        try:
            if word == '0':
                detected = '0'
                break
            elif re.search(f'(^|\s+)({alert})($|\s+)',word) != None:
                detected = alert
                break
            else:
                detected = np.nan
        except:
            detected = np.nan
        if detected != None: 
            break
    return detected
```

```python
# Detect special words and observe
df_focus.loc[:,'special_words'] = df_focus.water_tech2.apply(special_detect)
df_focus2 = df_focus[~pd.isnull(df_focus.special_words)][['water_tech2','special_words']].reset_index(drop=True)
print(f'Total dataset rows = {len(df_focus)}')
print(f'Total dataset rows (that contains special words) = {len(df_focus2)}')
```

### Focus on rows with special words (df_focus2)

```python
# Replace with np.nan
dict_replace = {'other': np.nan
               ,'0': np.nan}

def replace_in_dict(word):
    replaced = word
    if word in list(dict_replace.keys()):
        replaced = dict_replace[word]
    return replaced
```

### Clean 2: Remove entries that is in dict_replace

```python
df_focus.water_tech2 = df_focus.water_tech2.apply(replace_in_dict)
df_focus2.water_tech2 = df_focus2.water_tech2.apply(replace_in_dict)
(df_focus2['water_tech2'].value_counts()/len(df_focus2)).sort_values(ascending=False)
```

### Clean 3: Remove "none", "other"

```python
df_focus.loc[:,'water_tech2'] = df_focus.loc[:,'water_tech2'].str.replace(r'others*', '').str.replace(r'none', '').str.replace(r'\s+',' ').str.strip().replace(r'^\s*$', np.nan, regex=True)
df_focus2.loc[:,'water_tech2'] = df_focus2.loc[:,'water_tech2'].str.replace(r'others*', '').str.replace(r'none', '').str.replace(r'\s+',' ').str.strip().replace(r'^\s*$', np.nan, regex=True)
```

```python
# Observe entries that are left after cleaning
(df_focus2['water_tech2'].value_counts()/float(len(df_focus2.water_tech2.dropna())))\
.sort_values()\
.cumsum()\
.sort_values(ascending=False)\
.reset_index
```
